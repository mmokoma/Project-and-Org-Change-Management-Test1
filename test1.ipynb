<b>SIMPLE LINEAR REGRESSION</b>

$$Formula:     Y = bX + c$$

```python
#Import libraries 
from sklearn.linear_model import LinearRegression
```

```python
#Using reported cases in the last 24 hours to predict deaths in the next 24 hours
x = my_df[['Cases - newly reported in last 24 hours']] 
y = my_df['Deaths - newly reported in last 24 hours'] 
lm = LinearRegression() 
lm.fit(x,y)

Yhat=lm.predict(x)
Yhat[0:5]
```

```python
#Calculating the value of the intercept, c
lm.intercept_
```

```python
#Calculating the value of the slope, m
lm.coef_
```

```python
print('The value of R-squared is: ', lm.score(x, y))
```

```python
#import the function mean_squared_error from the module metrics
from sklearn.metrics import mean_squared_error
```

```python
y=lm.predict(x)
MSE = mean_squared_error(my_df['Deaths - newly reported in last 24 hours'], y)
print('The value of Mean Squared Error (MSE) is', MSE)
```

```python
#calculate RMSE
#import necessary libraries
from math import sqrt
```

```python
RMSE = sqrt(mean_squared_error(my_df['Deaths - newly reported in last 24 hours'], y))
print('The value of RMSE is' , RMSE)
```

<b>MULTI LINEAR REGRESSION</b>
$$Formula example:  Y = b1X1 + b2X2 + c$$

```python
# The multi variables here are; Cases - newly reported in last 7 days and Cases and Cases - newly reported in last 24 hours
multi_X = my_df[['Cases - newly reported in last 7 days','Cases - newly reported in last 24 hours']]
y1 = my_df['Deaths - newly reported in last 24 hours'] 
ml = LinearRegression() 
ml.fit(multi_X,y1)

Yhat1=ml.predict(multi_X)
Yhat1[0:5]
```

```python
#Calculating the value of the intercept, c
ml.intercept_
```

```python
#Calculating the value of the slope, m
ml.coef_
```

```python
print('The value of R-squared is: ', ml.score(multi_X, y1))
```

```python
y1=ml.predict(multi_X)
MSE1 = mean_squared_error(my_df['Deaths - newly reported in last 24 hours'], y1)
print('The value of Mean Squared Error (MSE) is', MSE1)
```

```python
RMSE1 = sqrt(mean_squared_error(my_df['Deaths - newly reported in last 24 hours'], y1))
print('The value of RMSE is' , RMSE1)
```
